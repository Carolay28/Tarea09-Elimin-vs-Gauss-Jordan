1. Para cada uno de los siguientes sistemas lineales, obtenga, de ser posible, una soluci√≥n con m√©todos gr√°ficos. Explique los resultados desde un punto de vista geom√©trico.

a) 

$ùë•_1 + 2ùë•_2 = 0,$

$x_1 ‚àí ùë•_2 = 0.$


```python
import numpy as np

A = np.array([
    [1, 2, 0],
    [1, -1, 0]
], dtype=float)

print("Matriz inicial:")
print(A)

print("\nResoluci√≥n:")
factor = A[1, 0] / A[0, 0]
print(f"\nR2 = R2 - ({factor}) * R1 \n")

A[1] = A[1] - factor * A[0]

print(A)

x2 = A[1, 2] / A[1, 1]
x1 = (A[0, 2] - A[0, 1] * x2) / A[0, 0]

print("\nSoluciones:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
```

b)

$x_1 + 2ùë•_2 = 3,$

$‚àí2ùë•_1 ‚àí 4ùë•_2 = 6.$

```python
import numpy as np

A = np.array([
    [1, 2, 3],
    [-2, -4, 6]
], dtype=float)

print("Matriz inicial:")
print(A)

print("\nResoluci√≥n:")

factor = A[1, 0] / A[0, 0]
print(f"\nR2 = R2 - ({factor}) * R1\n")

A[1] = A[1] - factor * A[0]

print(A)

if A[1,0] == 0 and A[1,1] == 0 and A[1,2] != 0:
    print("\nResultado:")
    print("0 = ", A[1,2])
    print("El sistema es inconsistente -> NO tiene soluci√≥n.")
else:
    x2 = A[1, 2] / A[1, 1]
    x1 = (A[0, 2] - A[0, 1] * x2) / A[0, 0]

    print("\nSoluciones:")
    print(f"x1 = {x1}")
    print(f"x2 = {x2}")
```

$2ùë•_1 + 4ùë•_2 ‚àí ùë•_3 = ‚àí1.$

```python
import numpy as np

A = np.array([
    [2, 1, 1, 1],
    [2, 4, -1, -1]
], dtype=float)

print("Matriz inicial:")
print(A)

print("\nResoluci√≥n:")

factor = A[1, 0] / A[0, 0]
print(f"\nR2 = R2 - ({factor}) * R1\n")

A[1] = A[1] - factor * A[0]

print(A)

coef = A[:, :3]
aug = A

rango_coef = np.linalg.matrix_rank(coef)
rango_aug = np.linalg.matrix_rank(aug)
num_incog = 3

if rango_coef < num_incog and rango_coef == rango_aug:
    print("\nEl sistema tiene infinitas soluciones.")
else:
    print("\nEl sistema tiene una √∫nica soluci√≥n o es incompatible.")
```

$-x_1 + 4x_2 + x_3 = 8,$

$5/3x_1 + 2/3x_2 + 2/3x_3 = 1,$

$x_1 + x_2 + 4x_3 = 11.$

```python
def r(x):
    return round(x, 2)

A = [
    [-1.0, 4.0, 1.0, 8.0],
    [5/3, 2/3, 2/3, 1.0],
    [1.0, 1.0, 4.0, 11.0]
]

print("Matriz inicial:")
for fila in A:
    print(fila)

print("\nResoluci√≥n\n")

m21 = r(A[1][0] / A[0][0])
print(f"m21 = {m21}")
A[1] = [r(A[1][j] - r(m21 * A[0][j])) for j in range(4)]

m31 = r(A[2][0] / A[0][0])
print(f"m31 = {m31}")
A[2] = [r(A[2][j] - r(m31 * A[0][j])) for j in range(4)]

print(" ")
for fila in A:
    print(fila)

print(" ")

m32 = r(A[2][1] / A[1][1])
print(f"m32 = {m32}")
A[2] = [r(A[2][j] - r(m32 * A[1][j])) for j in range(4)]

print(" ")
for fila in A:
    print(fila)

print("\nSustituci√≥n hacia atr√°s")

x3 = r(A[2][3] / A[2][2])
print(f"x3 = {x3}")

x2 = r((A[1][3] - r(A[1][2] * x3)) / A[1][1])
print(f"x2 = {x2}")

x1 = r((A[0][3] - r(A[0][1] * x2) - r(A[0][2] * x3)) / A[0][0])
print(f"x1 = {x1}")

print("\nSoluci√≥n con redondeo")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"x3 = {x3}")
```

```python
import numpy as np

A = np.array([
    [1/4, 1/5, 1/6, 9],
    [1/3, 1/4, 1/5, 8],
    [1/2, 1.0, 2.0, 8]
], dtype=np.float32)

print("Matriz inicial:")
print(A)

print("\nResoluci√≥n\n")

m21 = np.float32(A[1,0] / A[0,0])
m31 = np.float32(A[2,0] / A[0,0])

print(f"m21 = {m21}")
print(f"m31 = {m31}\n")

A[1] = np.float32(A[1] - m21*A[0])
A[2] = np.float32(A[2] - m31*A[0])

print(A, "\n")

m32 = np.float32(A[2,1] / A[1,1])
print(f"m32 = {m32}\n")

A[2] = np.float32(A[2] - m32*A[1])

print("Matriz final:")
print(A)

print("\nSustituci√≥n hacia atr√°s\n")

x3 = np.float32(A[2,3] / A[2,2])
print(f"x3 = {x3}")

x2 = np.float32((A[1,3] - A[1,2]*x3) / A[1,1])
print(f"x2 = {x2}")

x1 = np.float32((A[0,3] - A[0,2]*x3 - A[0,1]*x2) / A[0,0])
print(f"x1 = {x1}")

print("\nSoluci√≥n final:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"x3 = {x3}")
```

```python
import numpy as np

def revisar(a):
    
    A = np.array([
        [1, -1, a],
        [-1, 2, -a],
        [a, 1, 1]
    ], dtype=float)

    b = np.array([-2, 3, 2], dtype=float)

    Ab = np.column_stack((A, b))

    print(f"\nProbando con a = {a}")
    print("\nMatriz A =")
    print(A)
    print("\nMatriz Ab =")
    print(Ab)

    rango_A = np.linalg.matrix_rank(A)
    rango_Ab = np.linalg.matrix_rank(Ab)

    print("\nRango de A =", rango_A)
    print("Rango de Ab =", rango_Ab)

    if rango_A < 3 and rango_A == rango_Ab:
        print("\nEl sistema tiene infinitas soluciones.")
    else:
        print("\nEl sistema NO tiene infinitas soluciones.")

revisar(1)
revisar(-1)
```

```python
import numpy as np

A = np.array([
    [1,     1/2,   1/3,   1/4,   1/6],
    [1/2,   1/3,   1/4,   1/5,   1/7],
    [1/3,   1/4,   1/5,   1/6,   1/8],
    [1/4,   1/5,   1/6,   1/7,   1/9]
], dtype=np.float32)

print("Matriz inicial:")
print(A)

print("\nResoluci√≥n por Gauss-Jorda\n")

p1 = np.float32(A[0,0])
A[0] = np.float32(A[0] / p1)

m10 = np.float32(A[1,0])
m20 = np.float32(A[2,0])
m30 = np.float32(A[3,0])

A[1] = np.float32(A[1] - m10*A[0])
A[2] = np.float32(A[2] - m20*A[0])
A[3] = np.float32(A[3] - m30*A[0])

print(A, "\n")

p2 = np.float32(A[1,1])
A[1] = np.float32(A[1] / p2)

m01 = np.float32(A[0,1])
m21 = np.float32(A[2,1])
m31 = np.float32(A[3,1])

A[0] = np.float32(A[0] - m01*A[1])
A[2] = np.float32(A[2] - m21*A[1])
A[3] = np.float32(A[3] - m31*A[1])

print(A, "\n")

p3 = np.float32(A[2,2])
A[2] = np.float32(A[2] / p3)

m02 = np.float32(A[0,2])
m12 = np.float32(A[1,2])
m32 = np.float32(A[3,2])

A[0] = np.float32(A[0] - m02*A[2])
A[1] = np.float32(A[1] - m12*A[2])
A[3] = np.float32(A[3] - m32*A[2])

print(A, "\n")

p4 = np.float32(A[3,3])
A[3] = np.float32(A[3] / p4)

m03 = np.float32(A[0,3])
m13 = np.float32(A[1,3])
m23 = np.float32(A[2,3])

A[0] = np.float32(A[0] - m03*A[3])
A[1] = np.float32(A[1] - m13*A[3])
A[2] = np.float32(A[2] - m23*A[3])

print("Matriz final:")
print(A)

x1 = A[0,4]
x2 = A[1,4]
x3 = A[2,4]
x4 = A[3,4]

print("\nSoluci√≥n final:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"x3 = {x3}")
print(f"x4 = {x4}")
```
